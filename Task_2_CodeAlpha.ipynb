```python
# import libaries
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
#load data
Handwritten_data=datasets.load_digits()
```

```python
#Data allocation
len(Handwritten_data.data)
```

```python
#Data allocation
len(Handwritten_data.target)
```

```python
#Data Visualization by show Handwrittem Image
color_map = plt.cm.jet
plt.imshow(Handwritten_data.images[8], cmap=color_map, interpolation='nearest')
plt.title('Handwritten Image')

plt.show()
```

```python
color_map = plt.cm.jet # Define the color map for the images

num_images = 4

num_rows = 2
num_cols = num_images // num_rows

fig, axes = plt.subplots(num_rows, num_cols, figsize=(12,8))

image_index = 0

for row in range(num_rows):
    for col in range(num_cols):
        axes[row, col].imshow(Handwritten_data.images[image_index], cmap=color_map, interpolation='nearest')
        axes[row, col].set_title(f'Handwritten Image {image_index}')
        image_index += 1


plt.tight_layout() # Adjust the spacing between subplots

plt.show() # Display the plot
```

```python
Handwritten_data.target[4]
```

```python
#features allocation
X=Handwritten_data.data
Y=Handwritten_data.target
```

```python
#Data tranning & Testing
X_train ,X_test , Y_train ,Y_test=train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
# Calculate the length of X_train
len(X_train)
```

```python
# Calculate the length of X_test
len(X_test)
```

```python
# fit KNN Algorithm
knn=KNeighborsClassifier(n_neighbors=7)
```

```python
# Fit the k-nearest neighbors classifier to the training data
knn.fit(X_train,Y_train)
```

```python
# Make a prediction using the k-nearest neighbors classifier on a single test instance (index 359)
knn.predict(X_test[[359]])
```

```python
Y_test[359]
```

```python
# Print the accuracy of the k-nearest neighbors classifier on the test data
print(knn.score(X_test,Y_test))
```
